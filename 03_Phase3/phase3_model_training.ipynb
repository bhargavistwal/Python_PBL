# Phase 3 - Model Training

```python
# ============================================
# 1. IMPORT LIBRARIES
# ============================================

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import MinMaxScaler


# ============================================
# 2. LOAD DATA
# ============================================

df = pd.read_csv("cleaned_wifi_data.csv")

# Encode target
df["congestion_level"] = df["congestion_level"].map({
    "Low": 0,
    "Medium": 1,
    "High": 2
})


# ============================================
# 3. BALANCE DATASET (IMPORTANT)
# ============================================

print("\n--- BEFORE BALANCING ---\n")
print(df["congestion_level"].value_counts())

# Separate classes
df_low = df[df["congestion_level"] == 0]
df_med = df[df["congestion_level"] == 1]
df_high = df[df["congestion_level"] == 2]

# Find minimum class size
min_size = min(len(df_low), len(df_med), len(df_high))

# Downsample all to same size
df_balanced = pd.concat([
    df_low.sample(min_size, random_state=42),
    df_med.sample(min_size, random_state=42),
    df_high.sample(min_size, random_state=42)
])

# Shuffle dataset
df_balanced = df_balanced.sample(frac=1, random_state=42)

print("\n--- AFTER BALANCING ---\n")
print(df_balanced["congestion_level"].value_counts())


# ============================================
# 4. FEATURE & TARGET SPLIT
# ============================================

X = df_balanced.drop("congestion_level", axis=1)
y = df_balanced["congestion_level"]


# ============================================
# 5. SCALING
# ============================================

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

X_scaled = pd.DataFrame(X_scaled, columns=X.columns)


# ============================================
# 6. TRAIN TEST SPLIT
# ============================================

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)


# ============================================
# 7. MODEL TRAINING
# ============================================

model = LogisticRegression(max_iter=1000)

model.fit(X_train, y_train)


# ============================================
# 8. PREDICTION
# ============================================

y_pred = model.predict(X_test)


# ============================================
# 9. EVALUATION
# ============================================

print("\n--- RESULTS ---\n")

print("Accuracy:", accuracy_score(y_test, y_pred))

print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))


# ============================================
# 10. SAMPLE OUTPUT
# ============================================

print("\n--- SAMPLE PREDICTIONS ---\n")

for i in range(5):
    print("Actual:", y_test.iloc[i], "Predicted:", y_pred[i])
```
