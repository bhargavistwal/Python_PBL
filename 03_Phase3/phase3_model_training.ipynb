# Phase 3 - Model Training

```python
# =====================================================
# 1. IMPORT LIBRARIES
# =====================================================
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

# =====================================================
# 2. LOAD DATASET
# =====================================================
df = pd.read_csv("cleaned_wifi_data.csv")

print("\n--- DATA LOADED ---\n")
print(df.head())

# =====================================================
# 3. ENCODE TARGET
# =====================================================
df['congestion_level'] = df['congestion_level'].map({
    "Low": 0,
    "Medium": 1,
    "High": 2
})

# =====================================================
# 4. FEATURE & TARGET SPLIT
# =====================================================
X = df.drop("congestion_level", axis=1)
y = df["congestion_level"]

# =====================================================
# 5. SCALING
# =====================================================
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# =====================================================
# 6. TRAIN-TEST SPLIT
# =====================================================
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

print("\n--- TRAIN TEST SPLIT ---\n")
print("X_train:", len(X_train))
print("X_test:", len(X_test))

# =====================================================
# 7. MODEL TRAINING
# =====================================================

# Logistic Regression
lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_train, y_train)

# KNN
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)

# Decision Tree
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

print("\n--- MODELS TRAINED SUCCESSFULLY ---\n")

# =====================================================
# 8. PREDICTIONS
# =====================================================
y_pred_lr = lr_model.predict(X_test)
y_pred_knn = knn_model.predict(X_test)
y_pred_dt = dt_model.predict(X_test)

# =====================================================
# 9. SAMPLE OUTPUT
# =====================================================
print("\n--- SAMPLE PREDICTIONS ---\n")

for i in range(5):
    print(f"Actual: {y_test.iloc[i]} | LR: {y_pred_lr[i]} | KNN: {y_pred_knn[i]} | DT: {y_pred_dt[i]}")
```
