# Phase 5 - Model Manual Prediction

```python
# =====================================================
# 1. IMPORT LIBRARIES
# =====================================================
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression

# =====================================================
# 2. LOAD DATA
# =====================================================
df = pd.read_csv("cleaned_wifi_data.csv")

# =====================================================
# 3. ENCODE TARGET (SAFE)
# =====================================================
if df["congestion_level"].dtype == 'object':
    df["congestion_level"] = df["congestion_level"].map({
        "Low": 0,
        "Medium": 1,
        "High": 2
    })

# =====================================================
# 4. FEATURE & TARGET SPLIT
# =====================================================
X = df.drop("congestion_level", axis=1)
y = df["congestion_level"]

# =====================================================
# 5. SCALING
# =====================================================
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# =====================================================
# 6. TRAIN MODEL
# (Training here because this phase runs independently)
# =====================================================
model = LogisticRegression(max_iter=1000)
model.fit(X_scaled, y)

print("Model trained successfully!")

# =====================================================
# 7. MANUAL INPUT
# =====================================================
print("\n--- MANUAL TEST INPUT ---\n")

new_data = pd.DataFrame([{
    "num_devices": 25,
    "latency": 100,
    "packet_loss": 5,
    "signal_strength": -60,
    "bandwidth": 50,
    "channel": 6,
    "noise_level": -80,
    "throughput": 40,
    "upload_speed": 20,
    "download_speed": 60,
    "jitter": 10,
    "distance": 15,
    "obstacles": 2,
    "router_load": 10
}])

# Show input
print("--- INPUT DATA ---")
print(new_data)

# =====================================================
# 8. SCALE INPUT DATA
# =====================================================
new_data_scaled = scaler.transform(new_data)

# =====================================================
# 9. PREDICTION
# =====================================================
pred = model.predict(new_data_scaled)

# Convert to label
labels = {0: "Low", 1: "Medium", 2: "High"}

print("\n--- FINAL RESULT ---")
print("Predicted Congestion Level:", labels[pred[0]])

# =====================================================
# 10. INTERPRETATION (EXTRA MARKS)
# =====================================================
if pred[0] == 0:
    print("Low congestion → Network is smooth and stable")
elif pred[0] == 1:
    print("Medium congestion → Moderate traffic in network")
else:
    print("High congestion → Network is overloaded, performance may drop")
```
