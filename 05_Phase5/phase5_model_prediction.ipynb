# Phase 5 - Model Manual Prediction

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression

# Load data
df = pd.read_csv("cleaned_wifi_data.csv")

# Encode safely
if df["congestion_level"].dtype == 'object':
    df["congestion_level"] = df["congestion_level"].map({
        "Low": 0, "Medium": 1, "High": 2
    })

# Split
X = df.drop("congestion_level", axis=1)
y = df["congestion_level"]

# Scale
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Train model
model = LogisticRegression(max_iter=1000)
model.fit(X_scaled, y)

print("Model trained successfully!")

# ==============================
# MANUAL INPUT (FIXED)
# ==============================

print("\n--- MANUAL TEST ---\n")

# MUST match columns exactly
new_data = pd.DataFrame([{
    "num_devices": 25,
    "latency": 100,
    "packet_loss": 2,
    "signal_strength": -60,
    "bandwidth": 50,
    "channel": 6,
    "noise_level": -80,
    "throughput": 40,
    "upload_speed": 20,
    "download_speed": 60,
    "jitter": 10,
    "distance": 15,
    "obstacles": 2,
    "router_load": 10
}])

# Scale
new_data_scaled = scaler.transform(new_data)

# Predict
pred = model.predict(new_data_scaled)

labels = {0: "Low", 1: "Medium", 2: "High"}

print("Prediction:", labels[pred[0]])
```
