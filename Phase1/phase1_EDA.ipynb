## Wi-Fi Network Congestion Prediction - Phase 1

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1- creating simple dataset
data = {
    "num_devices": [10, 20, 30, 40, 50, 60],
    "latency": [20, 35, 50, 70, 90, 110],
    "packet_loss": [1, 2, 2.5, 3, 4, 5],
    "bandwidth": [50, 60, 65, 70, 75, 80],
    "load": [30, 45, 60, 75, 85, 95]
}

df = pd.DataFrame(data)

print("Dataset:")
print(df)

# 2- info
print("\nInfo:")
print(df.info())

# 3- missing values
print("\nMissing values:")
print(df.isnull().sum())

# 4- summary
print("\nSummary:")
print(df.describe())

# 5- graph 1
plt.hist(df["latency"])
plt.title("Latency Distribution")
plt.show()

# 6- graph 2
plt.scatter(df["num_devices"], df["latency"])
plt.xlabel("Devices")
plt.ylabel("Latency")
plt.title("Devices vs Latency")
plt.show()

 
```
