#  Phase 3 - Model training


```python
# ============================================
# 1. IMPORT LIBRARIES
# ============================================

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import MinMaxScaler


# ============================================
# 2. LOAD DATA
# ============================================

df = pd.read_csv("cleaned_wifi_data.csv")

print("\n--- DATA LOADED ---\n")
print(df.head())


# ============================================
# 3. ENCODING TARGET
# ============================================

df["congestion_level"] = df["congestion_level"].map({
    "Low": 0,
    "Medium": 1,
    "High": 2
})


# ============================================
# 4. FEATURE & TARGET SPLIT
# ============================================

X = df.drop("congestion_level", axis=1)
y = df["congestion_level"]


# ============================================
# 5. SCALING (VERY IMPORTANT)
# ============================================

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# convert back to DataFrame (important for indexing)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)


# ============================================
# 6. TRAIN TEST SPLIT (CORRECT)
# ============================================

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, shuffle=True
)

print("\n--- TRAIN TEST SPLIT ---\n")
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)


# ============================================
# 7. MODEL TRAINING
# ============================================

model = LogisticRegression(max_iter=1000)

model.fit(X_train, y_train)


# ============================================
# 8. PREDICTION
# ============================================

y_pred = model.predict(X_test)


# ============================================
# 9. EVALUATION
# ============================================

print("\n--- RESULTS ---\n")

print("Accuracy:", accuracy_score(y_test, y_pred))

print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))


# ============================================
# 10. SAMPLE OUTPUT
# ============================================

print("\n--- SAMPLE PREDICTIONS ---\n")

for i in range(5):
    print("Actual:", y_test.iloc[i], "Predicted:", y_pred[i])
```
