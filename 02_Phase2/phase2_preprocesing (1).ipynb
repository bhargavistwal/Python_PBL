# Phase 2 - Model Preprocessing


```python
# =====================================================
# 1. IMPORT LIBRARIES
# =====================================================
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split

# =====================================================
# 2. LOAD CLEANED DATASET
# =====================================================
df = pd.read_csv("cleaned_wifi_data.csv")

print("\n--- DATA LOADED ---\n")
print(df.head())

# =====================================================
# 3. VERIFY DATA CLEANLINESS
# =====================================================
print("\n--- MISSING VALUES ---\n")
print(df.isnull().sum())

print("\n--- DUPLICATES ---\n")
print(df.duplicated().sum())

# =====================================================
# 4. REMOVE DUPLICATES (IF ANY)
# =====================================================
df = df.drop_duplicates()

print("\n--- AFTER REMOVING DUPLICATES ---\n")
print(df.shape)

# =====================================================
# 5. ENCODING (TARGET VARIABLE)
# =====================================================
print("\n--- BEFORE ENCODING ---\n")
print(df['congestion_level'].unique())

df['congestion_level'] = df['congestion_level'].map({
    'Low': 0,
    'Medium': 1,
    'High': 2
})

print("\n--- AFTER ENCODING ---\n")
print(df['congestion_level'].unique())

# =====================================================
# EXTRA: TARGET DISTRIBUTION CHECK
# =====================================================
print("\nTarget Distribution:")
print(df['congestion_level'].value_counts())

# =====================================================
# 6. FEATURE & TARGET SPLIT
# =====================================================
X = df.drop("congestion_level", axis=1)
y = df["congestion_level"]

print("\n--- FEATURES SHAPE ---\n", X.shape)
print("\n--- TARGET SHAPE ---\n", y.shape)

# =====================================================
# EXTRA: FEATURE IMPORTANCE (CORRELATION WITH TARGET)
# =====================================================
temp_df = X.copy()
temp_df['target'] = y

print("\nFeature Importance (Correlation with Target):")
print(temp_df.corr(numeric_only=True)['target'].sort_values(ascending=False))

# =====================================================
# 7. MINMAX SCALING
# =====================================================
# Using MinMax Scaling for model training
minmax = MinMaxScaler()

X_minmax = minmax.fit_transform(X)
X_minmax = pd.DataFrame(X_minmax, columns=X.columns)

print("\n--- MINMAX SCALED DATA ---\n")
print(X_minmax.head())

# =====================================================
# EXTRA: SAVE PROCESSED DATA
# =====================================================
X_minmax.to_csv("processed_X.csv", index=False)
y.to_csv("processed_y.csv", index=False)

# =====================================================
# 8. STANDARD SCALING
# =====================================================
# Standard Scaling used for comparison
standard = StandardScaler()

X_standard = standard.fit_transform(X)
X_standard = pd.DataFrame(X_standard, columns=X.columns)

print("\n--- STANDARD SCALED DATA ---\n")
print(X_standard.head())

# =====================================================
# 9. TRAIN-TEST SPLIT
# =====================================================
# 80% training and 20% testing for better model generalization
X_train, X_test, y_train, y_test = train_test_split(
    X_minmax, y, test_size=0.2, random_state=42
)

print("\n--- TRAIN TEST SPLIT ---\n")
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)

# =====================================================
# 10. FINAL CHECK
# =====================================================
print("\n--- FINAL DATA READY FOR MODEL TRAINING ---\n")
print(X_train.head())
```
